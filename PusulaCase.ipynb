## TedaviAdi kolonunu ele alalim ve gorsellestirip eksik degerlerı dolduralım

```python
df["TedaviAdi"].unique()
```

```python
df["TedaviAdi"] = df["TedaviAdi"].str.replace(r"\+", " ve ", regex=True)
df["TedaviAdi"] = df["TedaviAdi"].str.replace(r"-", " ", regex=True)
df["TedaviAdi"] = df["TedaviAdi"].str.replace(r"\s+", " ", regex=True).str.strip()
df["TedaviAdi"] = df["TedaviAdi"].str.lower()
```

```python
df["TedaviAdi"].unique()
```

```python
replace_dict = {
    "ağrsıı": "ağrısı",
    "i̇mpingement": "impingement",
    "i̇mplantı": "implantı",
    "bialetarl": "bilateral",
    "muscular strain": "muskuler strain",
    "ağrisi": "ağrısı"
}
df["TedaviAdi"] = df["TedaviAdi"].replace(replace_dict, regex=True)
```

```python
def group_treatment(x):
    if "dorsalji" in x:
        return "dorsalji"
    elif "gonartroz" in x or "menisküs" in x:
        return "gonartroz / menisküs"
    elif "aşil" in x:
        return "aşil"
    elif "kalça implant" in x:
        return "kalça implantı"
    elif "omuz" in x:
        return "omuz"
    else:
        return x

df["TedaviAdi"] = df["TedaviAdi"].apply(group_treatment)
```

```python
def clean_text(text):
    if pd.isna(text):
        return np.nan
    text = text.lower()
    text = text.replace('\xa0', ' ')
    text = text.replace('_', ' ')
    text = text.replace(',,', ',')
    text = text.strip()
    return text

df['Tanilar'] = df['Tanilar'].apply(clean_text)
```

```python
df["Tanilar"].unique()
```

```python
def simplify_diagnosis(x):
    x = str(x).lower()
    if 'dorsal' in x or 'boyun ağrısı' in x:
        return 'Sırt/Boyun Ağrısı'
    elif 'omuz' in x or 'rotator' in x or 'biseps' in x:
        return 'Omuz/Üst Kol Sorunları'
    elif 'menisküs' in x or 'gonartroz' in x or 'diz' in x:
        return 'Diz Eklem Sorunları'
    elif 'koksartroz' in x or 'kalça' in x:
        return 'Kalça Eklem Sorunları'
    elif 'aşil' in x or 'plantar' in x or 'ayak' in x:
        return 'Ayak/Akil Tendon Sorunları'
    elif 'hipertansiyon' in x or 'kalp' in x or 'diabetes' in x:
        return 'Kardiyovasküler/Metabolik Hastalıklar'
    elif 'fibromiyalji' in x:
        return 'Genel Kas-İskelet Ağrısı'
    else:
        return 'Diğer'

df['Tanilar'] = df['Tanilar'].apply(simplify_diagnosis)
df['Tanilar'].unique()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Tanilar', order=df['Tanilar'].value_counts().index, palette='Set2')
plt.xticks(rotation=45, ha='right')
plt.title("Tanıların Sadeleştirilmiş Dağılımı")
plt.ylabel("Hasta Sayısı")
plt.xlabel("Tanı Grupları")
plt.tight_layout()
plt.show()
```

```python
le = LabelEncoder()
df['Tanilar'] = le.fit_transform(df['Tanilar'])
```

## TedaviAdi kolonun encode edelim

```python
df["TedaviAdi"].unique()
```

```python
def tedavi_grup(x):
    x = str(x).lower()  # küçük harfe çevir
    if 'el rehab' in x or 'rehabilitasyon' in x or 'el ' in x:
        return 'el_rehabilitasyonu'
    elif 'diz' in x or 'kalça' in x or 'implant' in x:
        return 'eklem_implant_reha'
    elif 'ayak' in x or 'ayak bileği' in x or 'metatars' in x:
        return 'ayak_bileği'
    elif 'kas' in x or 'ekstremite' in x or 'atrofi' in x:
        return 'kas_egzersiz'
    elif 'tendin' in x or 'tenosinovit' in x or 'tendon' in x:
        return 'tendinit_tendon'
    elif 'bel' in x or 'boyun' in x or 'spinal' in x:
        return 'spinal_boyun_bel'
    elif 'hemip' in x or 'paralizi' in x or 'pleji' in x:
        return 'hemipleji_paralizi'
    else:
        return 'diger'

df['TedaviAdi'] = df['TedaviAdi'].apply(tedavi_grup)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='TedaviAdi', order=df['TedaviAdi'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Tedavi Gruplarının Dağılımı')
plt.show()
```
